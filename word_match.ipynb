```python
def process(fp):
    with open(fp) as f:
        documents = json.load(f)
    for d in documents:
        token_num = 0
        d['SENTS'] = [s['text'] for s in d['SENTXS']]
        d['tokens'] = [bert_tokenizer.tokenize(s['text']) for s in d['SENTXS']]
        for s in d['tokens']:
            token_num += len(s)
        d['token_num'] = token_num
    token_num = 0
    for d in documents:
        token_num += d['token_num']
    print(token_num/len(documents))
    return documents
```

```python
documents = process(config.FGC_TEST)
```

```python
for d in documents:
    for q in d['QUESTIONS']:
        q['tokens'] = bert_tokenizer.tokenize(q['QTEXT'])
```
